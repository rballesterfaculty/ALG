## 2

Any convex polygon can be written as the set of 2D points that satisfies a set of linear constraints (one constraint per edge). Linear constraints mean the problem "find the closest points between two convex polygons" can be reduced to quadratic programming.

If the polygon is not convex, linear constraints no longer work: there will always be a subset of the polygon that violates at least one constraint.

## 3

If we simply loop through all $n$ edges, this takes $n$ line-segment intersection tests, which will cost $O(n)$ time.

For a more efficient solution, note that ray casting may be performed with any orientation of the ray, so we can choose a horizontal ray towards the right. Then, to test whether a point $(x, y)$ belongs to the polygon, we must find all edges $e_i = (x_i, y_i, x_{i+1}, y_{i+1})$ such that:

- Either $x_i > x$ or $x_{i+1} > x$ (or both);
- and either $y_i < y < y_{i+1}$, or $y_{i+1} < y < y$

(these conditions are not sufficient, but they are necessary to cross the horizontal ray departing from $x, y$ towards the right)

To find all candidate edges, we treat each edge as a 4D point $(x_i, y_i, x_{i+1}, y_{i+1})$, insert all edges in a nested BST (can be a KD-tree, which we saw in another session, or another structure, like a 4D octree) and perform range queries (each of which will have cost $O(\log n)$).

## 4

We can dilate the points (using a constant number of edges to approximate each resulting circle as a polygon) by $d$. Then, we can check for intersections between the polygons and identify those pairs that just "touch" each other.

*Note: approximating the circles by polygons leads to some inaccuracy, so we can use the following criterion: if polygons $A$ and $B$ intersect in a way that no edge of $A$ is completely contained in $B$, then $A$ and $B$ just touch each other.*

## 5

Let $\mathcal{S}$ be the set of points, and suppose $A, B \in \mathcal{S}$ are the pair of points that are the farthest apart. Let $d$ be the distance from $A$ to $B$. Draw two circles of radius $d$, one centered on each point. Note that $\mathcal{S} \subseteq \text{ConvexHull}(\mathcal{S})$ and that $\mathcal{S}$ must be completely enclosed in the area where both circles overlap. Last, $A$ and $B$ cannot belong to the interior of an edge of $\text{ConvexHull}(\mathcal{S})$, because then part of the convex hull would fall outside the overlap area. Therefore, $A$ and $B$ must be vertices of $\text{ConvexHull}(\mathcal{S})$.
