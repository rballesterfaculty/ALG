## 5



```python
def exponentiate(A, n):
    if n == 1:  # Base case
        return A
    elif n % 2 == 1:
        return A @ exponentiate(A, n-1)
    else:
        part = exponentiate(A, n // 2)
        return part @ part

import numpy as np
A = np.array([[1, 2], [3, 4]])
n = 1000000
print(exponentiate(np.eye(2) + A/n, n))
```

## 6

The recurrence is:

$$\begin{pmatrix}F_{n} \\ F{n+1}\end{pmatrix} = \begin{pmatrix}
0 & 1 \\
1 & 1
\end{pmatrix} \cdot \begin{pmatrix}F_{n-1} \\ F{n}\end{pmatrix}$$

In Python:

```python
A = np.array([[0, 1], [1, 1]])
u = np.array([0, 1])
result = exponentiate(A, 20)
print((result @ u)[0])
```

```python
def carry(u):
    u = u.tolist()[::-1]  # Reverse so that we can append at the end. We'll do the carry left-to-right
    i = 0
    while True:
        if u[i] >= 10:
            carryover = u[i] // 10
            # u[i-1] += u[i] // 10
            u[i] = u[i] % 10
        else:
            carryover = 0
        if i < len(u)-1:
            u[i+1] += carryover
        else:
            if carryover > 0:
                u.append(carryover)
            if carryover < 10:
                break
        i += 1
    return u[::-1]


print(carry(np.convolve(p, q)))

# which corresponds to

print(321*654)
```
