## 1

1a. The servers will have tasks $(p_1, p_4)$, $(p_2, p_5)$, and $(p_3, p_6)$, respectively.

1b. Take loads $p_1 = 1, p_2 = 2, p_3 = 3$. The greedy strategy will assign $p_1$ and $p_3$ (total workload = 4) to the first server and $p_2$ to the second server. An optimal strategy would send $p_1, p_2$ to the first and $p_3$ to the second (equal balance).

1c. Loads $p_1 = 1, p_2 = 1, p_3 = 2$ do the trick: $\text{cost}_{\text{greedy}} = 3$, while $\text{cost}_{\text{OPT}} = 2$. Therefore, the competitive ratio is $3 / 2 = 1.5$ which is the highest possible at $m = 2$ (since the upper bound is $2 - 1/m$).



## 2

2a. $\text{cost} = 5$. The ratio is $\text{cost} / \text{OPT} = 5 / 5 = 1$.

2b. For a strategy that decides to buy on day $N+1$, the worst case always happens when the number of skiing days is precisely $N+1$. That scenario makes the player rent for $N$ days, and then buy a ski set that will be used only one day. The cost using that strategy would be $N + B$, while the optimal cost would be $\min(B, N+1)$. The competitive ratio is attained at $\max_N \frac{N + B}{\min(B, N+1)}$ which cannot be less than 2.
